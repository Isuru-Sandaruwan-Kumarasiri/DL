```python
import pandas as pd 
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load Dataset
data = pd.read_csv("D:\DL\env\datasets\sonar.csv",header=None)
data=data.values
```

```python
data[:1]
```

```python
data.shape
```

```python
## divide dataset into x and y
X = data[:,0:60].astype(float)
Y = data[:,60]
```

```python
# encode labels 
encorder = LabelEncoder()
encorder.fit(Y)
Y = encorder.transform(Y)
```

```python
# scale the features
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
### Divide dataset into train and test
from sklearn.model_selection import train_test_split    
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
tf.random.set_seed(42)

# Build the model with Dropout layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30,activation='relu',input_shape=(60,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(30,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
# create the model 
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])
```

```python
# train the model
epoch_number = 100
history = model.fit(X_train,Y_train,epochs=epoch_number)
```

```python
# evaluate the model
model.evaluate(X_test,Y_test)
```
